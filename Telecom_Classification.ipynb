# Spliting the dataset

- Our aim is to work only with training dataset and testing on the testing dataset(later)

```python
x=data.drop(['Churn'],axis=1)
y=data['Churn']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

## Stratified shuffle shift

```python
from sklearn.model_selection import StratifiedShuffleSplit

split=StratifiedShuffleSplit(n_splits=7,test_size=0.3,random_state=26)
for train_index, test_index in split.split(data, data["Churn"]):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]
```

```python
strat_train_set.head()
```

### Error comparison

```python
def churn_rate(df):
    return data["Churn"].value_counts() / len(df)

train_set, test_set = train_test_split(data,test_size=0.2,random_state=23)

compare_props = pd.DataFrame({
    "Overall": churn_rate(data),
    "Stratified": churn_rate(strat_test_set),
    "Random": churn_rate(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100
```

# Model testing 

```python
new_df_train = strat_train_set.drop(['Churn'], axis=1)
new_df_label = strat_train_set['Churn']
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(new_df_train, new_df_label)
```

```python
from sklearn.metrics import mean_squared_error

churn_pred=log_reg.predict(new_df_train)
log_mse = mean_squared_error(new_df_label, churn_pred)
log_rmse = np.sqrt(log_mse)
log_rmse
```

```python
churn_pred
```

```python
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(new_df_label, churn_pred)
lin_mae
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_reg = DecisionTreeClassifier(random_state=42)
tree_reg.fit(new_df_train, new_df_label)
```

```python
churn_pred2 = tree_reg.predict(new_df_train)
tree_mse = mean_squared_error(new_df_label,churn_pred2)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

## Cross-checking

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, new_df_train, new_df_label,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```

```python
scores = cross_val_score(log_reg, new_df_train, new_df_label,
                         scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(lin_rmse_scores)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_reg = RandomForestClassifier(random_state=42)
forest_reg.fit(new_df_train, new_df_label) 

scores = cross_val_score(forest_reg, new_df_train, new_df_label,
                         scoring="neg_mean_squared_error", cv=10)
for_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(for_rmse_scores)
```

```python
churn_pred3 = forest_reg.predict(new_df_train)
for_mse = mean_squared_error(new_df_label,churn_pred3)
for_rmse = np.sqrt(for_mse)
for_rmse
```

# Selecting best parameters

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

forest_reg = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(new_df_train, new_df_label)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
pd.DataFrame(grid_search.cv_results_)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=8),
    }

forest_reg = RandomForestClassifier(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(new_df_train, new_df_label)
```

```python
cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```
